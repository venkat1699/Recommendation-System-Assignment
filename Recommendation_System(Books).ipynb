```python
import numpy as np
import pandas as pd
```

```python
surya = pd.read_csv('book rec.csv',encoding ='latin-1')
surya
```

```python
surya_=surya.sort_values(by='User.ID').reset_index(drop = True)
surya_
```

```python
surya_df=surya_.drop(['Unnamed: 0'],axis=1)
surya_df
```

```python
surya_df=surya_df.rename({'User.ID':'userid','Book.Title':'booktitle','Book.Rating':'Rating'},axis=1)
surya_df
```

```python
surya_df.nunique()
```

```python
deva_df= surya_df.pivot_table(index='userid',
                        columns='booktitle',
                        values='Rating').reset_index(drop=True)
```

```python
deva_df
```

```python
deva_df.index = surya_df.userid.unique()
deva_df
```

```python
deva_df.fillna(0,inplace=True)
```

```python
deva_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
deva = 1-pairwise_distances(deva_df.values,metric ='cosine')
```

```python
deva
```

```python
deva1 = pd.DataFrame(deva)
```
